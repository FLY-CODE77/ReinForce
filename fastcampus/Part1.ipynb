## Reinforcement

- 주어진 상황이 있을 때, 가장 좋은 행동을 찾기위해 학습하는것

# 강화학습 문제의 특징

- 정답을 모른다! 대신 행동에 대한 보상/ 다음 상태가 주어진다
- 현재의 의사결정이 미래에 영향을 미친다.
- 문제의 구조를 모른다.(강화 학습 에이전트는 수 많은 상호작용으로 서울에서 대전으로 이동시 보상과 결과를 알수있다)

# 강화학습의 구성요소

- Agent
- Enviroment

- 에이전트는 현제 상태 $s_t$를 관측하고, 현재 상태를 이용해 적합한 행동 $a_t$를 구한 후 그를 환경에 가한다.
- 환경은 행동을 반영하고, 그로 인한 다음 상태 $s_{t+1}$ 및 보상 $r_{t+1}$를 에이전트에게 전달한다

# - 보상(reward) & 상태(State)
- 보상은 하나의 숫자로 표현된 행동에 대한 평가 지표
- 상태는 현재 상황을 잘 기술하는 정보 

# Agent 
- 정책 policy : Agent 가 특정 상태에서 행동을 하는 방식을 결정짓는 요소
- 가치함수 Value function : 각각의 상태/ 행동이 얼마나 좋을지 평가하는 요소
- 모델 : Agent가 추측하는 세상

- 일반적으로 강화학습은 모델이 없는 강화학습을 지칭한다

# 확률 변수(random varianle) 과 확률 사건(event)

- 이산 확률 분포(discrete probability distribution)

- 주사위를 한번 던지면 각 눈이 나올 확률은 1/6

- 확률 공리
    - 모든 사건의 확률은 항상 0보다 크거나 같다
    - 표본 공간은 1.0 이다
    - 배반사건(교집합이 없는) 확률은 각각의 확률의 합과 같다
    
- 조건부 확률(conditional probabiltiy)
- P(y|x) = p(y 교집합 x)/ p(x)
- 어떤사건이 일어났다고 가정했을때 다른 사건이 일어날 확률


- 기댓값(expectation val)
- 확률 변수 x 의 평균을 계산한것


- 기댓값을 근사적으로 계산하는 방법:monte-carlo
- 확률을 모르는 경우 정말 많이 던져서 나온 값들의 평균을 구해준다


- 조건부 기댓값(Conditional expectation)
- E(Y|X = x) = $\sum_ip(Y=y_i|X=x)y_i$
